```r
library(repr)
library(tidyverse)
library(tidymodels)
```

```r
url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
downloaded_file <- download.file(url, destfile = "cleveland.data")
cleveland <- read_csv("cleveland.data", col_names = FALSE)
```

```r
colnames(cleveland) <- c("age", "sex", "cp", "trestbps", "chol", "fbs", "restecg",
                         "thalach", "exang", "oldpeak", "slope", "ca", "thal", "num")
cleveland
```

```r
#change to the na.rm method
# removing rows which have NA values
cleveland_no_na <- cleveland |>
    na.omit()

# selecting the columns which we will be using in our analysis
cleveland_selected <- cleveland_no_na |>
    select(sex, cp, trestbps, thalach)

#should give credits
#https://dplyr.tidyverse.org/reference/if_else.html
# mutating the data into human readable
cleveland_mutated <- cleveland_selected |>
    mutate(sex_chr = (if_else(sex == 1, "male", "female")),
            cp_chr = (if_else(cp == 1, "typical angina", 
                             if_else(cp == 2, "atypical angina",
                                    if_else(cp == 3, "non-anginal pain", "asymptomatic")))),
          trestbps_scaled = scale(trestbps),
          thalach_scaled = scale(thalach))

cleveland_mutated
```
