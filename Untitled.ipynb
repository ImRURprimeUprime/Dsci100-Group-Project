```r
library(repr)
library(tidyverse)
library(tidymodels)
```

```r
url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
downloaded_file <- download.file(url, destfile = "cleveland.data")
cleveland <- read_csv("cleveland.data", col_names = FALSE)
```

```r
colnames(cleveland) <- c("age", "sex", "cp", "trestbps", "chol", "fbs", "restecg",
                         "thalach", "exang", "oldpeak", "slope", "ca", "thal", "num")
```

```r
#change to the na.rm method
# removing rows which have NA values
cleveland_no_na <- cleveland |>
    na.omit()

#should give credits
#https://dplyr.tidyverse.org/reference/if_else.html
# mutating the data into human readable
cleveland_mutated <- cleveland_no_na |>
    mutate(sex_chr = (if_else(sex == 1, "male", "female")),
            cp_chr = (if_else(cp == 1, "typical angina", 
                             if_else(cp == 2, "atypical angina",
                                    if_else(cp == 3, "non-anginal pain", "asymptomatic")))),
          age_standardized = scale(age, center = TRUE, scale = TRUE),
          thalach_standardized = scale(thalach, center = TRUE, scale = TRUE))

# selecting the columns which we will be using in our analysis
cleveland_selected <- cleveland_mutated |>
    select(sex_chr, cp_chr, age_standardized, thalach_standardized) 


# splitting the dataset into training set and testing set, produce the training set table
cleveland_split <- initial_split(cleveland_selected, prop = 0.75, strata = cp_chr)
cleveland_train <- training(cleveland_split)
cleveland_test <- testing(cleveland_split)
cleveland_train

# produce a graph for the training set
options(repr.plot.width=13, repr.plot.height=13)
cleveland_plot <- cleveland_selected |>
    ggplot(aes(x = age_standardized, y = thalach_standardized)) +
    geom_point(aes(color = cp_chr, shape = sex_chr)) +
    labs(x = "Age (standardized)", y = "Maximum Heart Rate Achieved (standardized)",
         color = "Type Of Chest Pain", shape = "Sex") +
    theme(text = element_text(size = 20))

cleveland_plot
```
