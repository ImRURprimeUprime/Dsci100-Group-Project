```r
library(repr)
library(tidyverse)
library(tidymodels)
```

```r
url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
downloaded_file <- download.file(url, destfile = "cleveland.data")
cleveland <- read_csv("cleveland.data", col_names = FALSE)
```

```r
colnames(cleveland) <- c("age", "sex", "cp", "trestbps", "chol", "fbs", "restecg",
                         "thalach", "exang", "oldpeak", "slope", "ca", "thal", "num")
#produce df with headers
cleveland
```

```r
#change to the na.rm method
# removing rows which have NA values that are indicated by ?
cleveland_no_missing <- cleveland 


cleveland_mutated <- cleveland_no_missing |>
    mutate(diagnosis = (if_else(num == 0, "not diagnosed", "diagnosed")),
           age_standardized = scale(age, center = TRUE, scale = TRUE),
           chol_standardized = scale(chol, center = TRUE, scale = TRUE),
           thalach_standardized = scale(thalach, center = TRUE, scale = TRUE))

# selecting the columns which we will be using in our analysis
cleveland_selected <- cleveland_mutated |>
    select(diagnosis, age_standardized, chol_standardized, thalach_standardized)

# splitting the dataset into training set and testing set, produce the training set table
cleveland_split <- initial_split(cleveland_selected, prop = 0.75, strata = diagnosis)
cleveland_train <- training(cleveland_split)
cleveland_test <- testing(cleveland_split)
cleveland_train

# produce scatterplot from training set
heart_rate_scatterplot <- cleveland_selected |>
    ggplot(aes(x = age_standardized, y = thalach_standardized, z = chol_standardized)) +
    geom_point(aes(color = diagnosis)) +
    labs(x = "Age (standardized)", y = "Maximum Heart Rate Achieved (standardized)",
         color = "Diagnosis") +
    theme(text = element_text(size = 20)) +
    ggtitle("Maximum Heart Rate vs Age With Diagnosis Indicated By Colour")
# produce a barplot based on the training set
heart_rate_barplot <- cleveland_selected |>
    ggplot(aes(x = diagnosis)) +
    geom_bar()+
    labs(y = "Number of People") +
    theme(text = element_text(size = 20)) +
    ggtitle("Number of People Diagnosed With Heart Disease vs Not Diagnosed")

heart_rate_barplot
heart_rate_scatterplot
```
